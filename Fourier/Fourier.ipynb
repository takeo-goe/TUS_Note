# フーリエ級数展開

次の周期関数のフーリエ級数展開を考える。

$$
f(t) =
\begin{cases}
0 & \text{if } -1 \le t < 0 \\
1 & \text{if } 0 \le t < 1
\end{cases},\qquad
f(t+2) = f(t)
$$


```python
fig, ax = plt.subplots(figsize=(13, 3))
plot_periodic_ex1(ax)
plt.tight_layout()
plt.show()
```

**Step1: フーリエ係数を計算する**

$$
C_n = \frac{1}{2L}\int_{-L}^{L} f(t)\exp\!\left[-i\frac{n\pi}{L} t\right]\,dt
$$

$$
C_n =
\begin{cases}
\displaystyle \frac12 & \text{if } n = 1,\\[4pt]
\displaystyle \frac{i}{2 n \pi}\left( (-1)^n - 1 \right) & \text{if } n \neq 1.
\end{cases}
$$


```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
plot_periodic_ex1(ax[0])
plot_Cn_ex1(ax[1])
plt.tight_layout()
plt.show()
```

係数 \(C_n\)：重ね合わせる波動の振幅と考えて良い。

**Step2: フーリエ級数展開を計算する**

$$
f(t)
= \sum_{n=-\infty}^{\infty} C_n \exp[i n \pi t]
\left(
= \frac{1}{2}
  + \sum_{k=1}^{\infty} \frac{2}{(2k+1)\pi}\sin[(2k+1)\pi t]
\right)
$$

実際にはコンピューターは \(\infty\) を扱えないので，
\(|n|\le N\) で総和を打ち切ってフーリエ級数展開を計算する：

$$
f(t) \approx \sum_{n=-N}^{N} C_n \exp[i n \pi t].
$$

\(N\) をどの程度まで大きくすれば，元の関数 \(f(t)\) を再現できるかを調べてみよう：


```python
N = 1
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
plot_Cn_ex1(ax[0], N, N)
plot_periodic_reconst_ex1(ax[1], N)
plt.tight_layout()
plt.show()
```

# フーリエ級数展開

次の周期関数のフーリエ級数展開を考える．

$$
\begin{aligned}
f(t) &= t \qquad (-1 \le t < 1) \\
f(t+2) &= f(t)
\end{aligned}
$$


```python
fig, ax = plt.subplots(figsize=(13, 3))
plot_periodic_ex2(ax)
plt.tight_layout()
plt.show()
```

$$
\begin{aligned}
C_n &= \frac{1}{2L}\int_{-L}^{L} f(t)\exp\!\left[-i\frac{n\pi}{L}t\right]\,dt \\[6pt]
&=
\begin{cases}
0 & \text{if } n = 1,\\[4pt]
\dfrac{i}{n\pi}(-1)^n & \text{if } n \neq 1
\end{cases}
\end{aligned}
$$

$$
f(t) = \sum_{n=-\infty}^{\infty} C_n \exp[i n \pi t]
\left(
= \sum_{n=1}^{\infty} \frac{2(-1)^{\,n+1}}{n\pi}\,\sin(n\pi t)
\right)
$$

# フーリエ変換

次の関数を考える．

$$
f(t) =
\begin{cases}
1 & \text{if } 0 \le t \le 1,\\
0 & \text{other}
\end{cases}
$$


```python
fig, ax = plt.subplots(figsize=(13, 3))
plot_non_periodic_ex1(ax, t = np.linspace(-2, 6, 1000))
plt.tight_layout()
plt.show()
```

**Step1: フーリエ変換を行い，関数 \(f(t)\) を波数空間における関数 \(F(k)\) に変換する．**

$$
\begin{aligned}
F(k) &= \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} f(t)\exp[-ik t]\,dt \\[6pt]
     &= \frac{1 - e^{-ik}}{\sqrt{2\pi}\, i k}
\end{aligned}
$$


```python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
plot_non_periodic_ex1(ax[0], t=np.linspace(-2, 6, 1000))
plot_non_periodic_ft_ex1(ax[1], t=np.linspace(-15, 15, 1000))
plt.tight_layout()
plt.show()
```

フーリエ変換後の関数 \(F(k)\)：重ね合わせる波動の振幅と考えて良い。

**Step2: フーリエ逆変換によって元の関数 \(f(t)\) を再現してみる．**

$$
f(t) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} F(k)\exp[i k t]\,dk
$$

積分範囲を \(k \in [-\infty,\infty]\) から \(k \in [-N,N]\) に打ち切って計算を行ってみる：

$$
f(t) \approx \frac{1}{\sqrt{2\pi}} \int_{-N}^{N} F(k)\exp[i k t]\,dk
$$

このとき，\(N\) をどの程度まで大きくすれば次のように定義されていた元の関数 \(f(t)\) を再現できるかを調べてみよう：

$$
f(t) =
\begin{cases}
1 & \text{if } 0 \le t \le 1,\\
0 & \text{other}
\end{cases}
$$


```python
N = 1
fig, ax = plt.subplots(1, 2, figsize=(16, 4))
D = 2 * N if N > 100 else 100
plot_non_periodic_ft_ex1(ax[0], t=np.linspace(-1 * D, D, 1000), range_color_red=(-N, N))
plot_non_periodic_invft_ex1(ax[1], N, t=np.linspace(-1, 4, 1000))
plt.tight_layout()
plt.show()
```
